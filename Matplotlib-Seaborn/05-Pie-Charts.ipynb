# Pie Charts

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### `matplotlib.pyplot.pie()`

Syntax: `matplotlib.pyplot.pie(x_data, labels, colors, startangle, counterclock, wedgeprops)`

Plot a pie chart. This function requires that the data be in a summarized form: the primary argument to the function will be the wedge sizes.

### `matplotlib.pyplot.axis()`

Syntax: `matplotlib.pyplot.axis(*args, emit=True, **kwargs)`

Convenience method to get or set some axis properties.

### Plot a simple Pie chart

```python
# read the pokemon dataset
pokemon = pd.read_csv("data/pokemon.csv")
pokemon.head()
```

```python
# make sorted count Series from generation_id column
sorted_counts = pokemon['generation_id'].value_counts()
sorted_counts
```

```python
# plot a pie chart
plt.pie(sorted_counts, labels=sorted_counts.index, startangle=90, counterclock=False)
plt.axis('square');
```

```python
# prepare labels
labels = ['gen5', 'gen1', 'gen3', 'gen4', 'gen2', 'gen7', 'gen6']
explode = (.1, 0, 0, 0, 0, 0, 0)  # only explode the 1st slice (i.e. 'gen5')

# plot a pie chart
plt.pie(sorted_counts, labels=labels, startangle=90, counterclock=False,
        shadow=True,autopct='%1.1f%%', explode=explode)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

### Plot a simple Donut plot

```python
plt.pie(sorted_counts, labels=sorted_counts.index, startangle=90, counterclock=False,
        wedgeprops={'width':.4});
```
